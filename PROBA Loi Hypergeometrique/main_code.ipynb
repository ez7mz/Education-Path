```python
import numpy as np       # Bibliothèque pour la manupilation des arrays
import time              # Bibliothèque pour la gestion de temp
import random            # Bibliothèque contien des génerateurs des valeur aléatoire
from os import system    # Bibliothèque Pour les commandes Système
import matplotlib.pyplot as plt  # Pour la visualisation des graphe
from scipy.special import comb

#======================= Fonctions ======================================================
# Fonction pour calculée la loi Hypergeometrique de X : 

def hypergeom(N, A, n, x):  # P( X = x )

    """
    N : taille de la population
    A : nombre total d'elements succes dans N
    n : nombre de tirages effectues e partir de N
    x : nombre d'elements succes dans notre tirage de n elements
    renvoie : La probabilite en X = x ---> P( X = x )
    """
    ACx = comb(A, x)  #Fonction pour calculer la compinaison pré definie en python
    NCx = comb(N - A, n - x)
    NCn = comb(N, n)

    return (ACx) * NCx / NCn

# Fonction pour calculée la loi Hypergeometrique de X <= x: 

def hypergeom_C(N, A, n, t, min_value=None):  # P( X <= x )

    """
    : param N : taille de la population
    : param A : nombre total d'elements succes dans N
    : param n : nombre de tirages effectues a partir de N
    : param t : nombre d'elements succes dans notre tirage de n elements jusqu'a t
    : renvoie : proba cumulee calcule jusqu'a t
    """
    if min_value: # si l'utilisateur entre une valeur minimale, nous commençons par sinon nous commençons par 0
        return np.sum([hypergeom(N, A, n, x) for x in range(min_value, t + 1)])

    return np.sum([hypergeom(N, A, n, x) for x in range(t + 1)])

# Fonction pour tracer la distrubition de X en fonction de P(X) :

def hypergeom_plot(N, A, n):

    """
    : param N: taille de la population
    : param A: nombre total d'elements succes dans N
    : param n: nombre de tirages effectues a partir de N
    """

    x = np.arange(0, n + 1)
    y = [hypergeom(N, A, n, x) for x in range(n + 1)]
    plt.plot(x, y, "o")
    plt.vlines(x, 0, y, lw=2)
    plt.xlabel("X")
    plt.ylabel("Probablite P(X)")
    plt.title("Distribution Hypergeometrique")
    plt.show()

# C'est juste une fonction pour quiter le programme :
def quite():
    print("\n" + "\t" * 3 + "Ok! Merci pour Votre Visite!...")
    time.sleep(1)
    system("CLS")

#===============================================================================================

# C'est la sumilation de L'exemple dans la presentation :

def Ampoules():
    result = []
    # en crée un list qui represente la lot contien 15 Ampoule (10 NonDefectues et 5 Defectues)
    lot = ["Amp" for x in range(10)] + ["AmpDef" for x in range(5)]
    random.shuffle(lot) # Pour mélanger les éléments de lot
    #C'est just des print de menu
    print("\n\n\n" + "=" * 48 + " Bonjour " + "=" * 48 + "\n\n")
    print("\t" * 3 + "# ", end="")
    print("[y] : Tirer 3 Ampoules ?\n")
    print("\t" * 3 + "# ", end="")
    print("[N] : pour voir example correction ?\n")
    print("\t" * 3 + "# ", end="")
    choix = input(" [ Y / N ] : ")

    if choix.lower() == "y":
        for i in range(3):
            draw = random.choice(lot) #en prende un élement aléatoire dans le lot
            result.append(draw) #en ajout cette élement a la liste des resultat
            lot.remove(draw) # remove cette element car ona un tirage sans remise
        print("\n" + "\t" * 3 + "# ", "votre resultat est : {}".format(result))
        print("\n" + "\t" * 3 + "# ", end="")
        choix = input("Show the Probability ? [ Y / N ] :")
        if choix.lower() == "y":
            print("\n" + "=" * 110)
            x = result.count("AmpDef") # calcule le nomber d'Ampoules Def dans le resultat pour calculée la probabilité
            print(
                "\t" * 3
                + "probabilité de resultat obtenu est : P(X = {}) = ".format(x),
                end="",
            )
            print(hypergeom(15, 5, 3, x))
            print("\n" + "=" * 110 + "\n")
            print("\n\n")
            choix = input("taper 1 pour ressayer et 0 pour quiter! : ")
            if int(choix) == 1:
                system("CLS")
                Ampoules()
            else:
                quite()
    elif choix.lower() == "n":   # C'est juste la Correction de l'exemple a l'aide de code
        print("\n" + "=" * 110 + "\n")
        choix = input("\n" + "\t" * 3 + "Show example correction ? [Y / N] : ")
        if choix.lower() == "y":
            print(
                "\n"
                + "\t" * 3
                + "probabilité de A ---> P(A) = {} \n".format(hypergeom_C(15, 5, 3, 1))
            )
            print(
                "\n"
                + "\t" * 3
                + "probabilité de B ---> P(B) = {} \n".format(hypergeom(15, 5, 3, 3))
            )
            print(
                "\n"
                + "\t" * 3
                + "probabilité de C ---> P(C) = {} \n".format(hypergeom(15, 5, 3, 1))
            )
            print("\n" + "=" * 110 + "\n")
            print("\n" + "=" * 110 + "\n")
            print("\n" + "\t" * 3 + "Esperance = {}".format(3 * (5 / 15)) + "\n")
            print(
                "\n"
                + "\t" * 3
                + "Variance = {}".format((12 / 14) * (1 - (5 / 15)))
                + "\n"
            )
            print("\n" + "=" * 110 + "\n")
            choix = input("\n" + "\t" * 3 + "Show distribution ? [Y / N ] : ")
            if choix.lower() == "y":
                hypergeom_plot(15, 5, 3)
                hypergeom_plot(1000, 5000, 200)
            else:
                quite()
        else:
            quite()
    else:
        print("\n" + "\t" * 3 + "Error Try again!...")
        time.sleep(1)
        system("CLS")
        Ampoules()

Ampoules()
        
print("programme run successfully with return value 0")
```
