# Serie Chronologique (Model Additif)

Une série chronologique est une variable statistique dont les observations sont repérées dans le temps.
Les séries chronologiques sont extrêmement utilisées dans les sciences sociales et, en particulier, en économie.

## Modèle multiplicatif et modèle additif

L’observation des séries chronologiques permet de distinguer deux grand types de série :
celles qui se conforment au modèle multiplicatif et celles qui se conforment au modèle
additif. Dans le modèle additif, les variations autour du trend demeurent dans une bande
de variation à peu près constante . Dans le modèle
multiplicatif, au contraire, les variations autour du trend s’amplifient.
